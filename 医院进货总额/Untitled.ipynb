## 数量取绝对值

```python
xiaoshou['Quantity'] = abs(xiaoshou['Quantity'])
```

## 日期取2017年

```python
xiaoshou = xiaoshou[xiaoshou['SalesDate'] >= pd.to_datetime('20170101')]
xiaoshou
```

## 填补缺失值

```python
jigou.fillna(0)
```

```python
jigou.loc[4,'Province'] = '辽宁省'
```

```python
jigou
```

```python
_ = xiaoshou.groupby(['InstitutionCode','SKU']).size().to_frame().rename(columns={0:'销量'})
_.head()
```

```python
_ = _.reset_index('SKU')
_
```

```python
_['SKU']=_['SKU'].astype(np.int)
```

```python
_['总额'] = 0
_
```

```python
def aa(tt):
    for index,row in jiage.iterrows():
        for i,j in _.iterrows():
            if j['SKU'] == row['SKU']:
                _.loc[i,'总额'] =  row['Price'] * _.loc[i,'销量']
                
    return tt
_ = _.apply(func=aa,axis=0)
```

```python
_.head()
```

## 计算出进货总额

```python
_ = _.groupby(['InstitutionCode']).sum().sort_values('总额',ascending=False)
_
```

## 合并到一张表格 

```python
jigou_index = jigou.set_index('InstitutionCode')
_ = _.join(jigou_index['Province'],on='InstitutionCode')
```

```python
_
```

### 无F6306，删除

```python
_ = _.drop(16)
```

```python
_
```

## 存为新的Excel

```python
writer = pd.ExcelWriter('new_data/Homework2.xlsx')

_.to_excel(
    writer,
    '各省医院进货金额排行',
    index = True
    
)
xiaoshou.to_excel(
    writer, # 写入数据
    'Sales', # 工作表标签
    index=False # 不存入行索引
)

jigou.to_excel(
    writer, # 写入数据
    'Institution', # 工作表标签
    index=False # 不存入行索引
)

chanpin.to_excel(
    writer, # 写入数据
    'Product', # 工作表标签
    index=False # 不存入行索引
)
jiage.to_excel(
    writer, # 写入数据
    'Price', # 工作表标签
    index=False # 不存入行索引
)

writer.save()
```
